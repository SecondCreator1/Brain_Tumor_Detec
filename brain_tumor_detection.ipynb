```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading images 

```python
# Paths to your training and testing datasets
training_path = 'dataset/training'
testing_path = 'dataset/testing'

# Folders representing each class
classes = ['glioma', 'meningioma', 'notumor', 'pituitary']
```

```python
# Function to count images in each class folder
def count_images(path):
    class_counts = {}
    for cls in classes:
        class_dir = os.path.join(path, cls)
        class_counts[cls] = len(os.listdir(class_dir))
    return class_counts
```

```python
# Get counts for training and testing sets
training_counts = count_images(training_path)
testing_counts = count_images(testing_path)
```

```python
# Print the results
print("Training Counts")
print(training_counts)

print("\nTesting Counts")
print(testing_counts)
```

```python
# Plot pie charts for training and testing sets
total_train = sum(training_counts.values())
total_test = sum(testing_counts.values())
```

```python
# Create pie chart for training data
plt.figure(figsize=(20, 7))
```

```python
# Subplot 1: Training Data
plt.subplot(1, 3, 1)
plt.pie(training_counts.values(), labels=training_counts.keys(), autopct=lambda p: '{:.2f}%\n{:,.0f}'.format(p, p * total_train / 100), 
        colors=['red', 'green', 'blue', 'orange'], startangle=90)
plt.title('Training Data')
```

```python
# Subplot 2: Train-Test Split
plt.subplot(1, 3, 2)
split = [total_train, total_test]
plt.pie(split, labels=['Train', 'Test'], autopct='%1.2f%%', colors=['teal', 'orange'], startangle=90)
plt.title('Train Test Split')
```

```python
# Subplot 3: Testing Data
plt.subplot(1, 3, 3)
plt.pie(testing_counts.values(), labels=testing_counts.keys(), autopct=lambda p: '{:.2f}%\n{:,.0f}'.format(p, p * total_test / 100), 
        colors=['red', 'green', 'blue', 'orange'], startangle=90)
plt.title('Testing Data')
```

```python
# Display the pie charts
plt.tight_layout()
plt.show()
```

A balanced distribution of categories in our training data is crucial for optimal model learning. It allows for comprehensive understanding of each category's characteristics, prevents biases, enhances generalization, and enables iterative refinement, leading to improved performance in accurately categorizing new data.

We also have a nice split percentage in our training and testing set.
